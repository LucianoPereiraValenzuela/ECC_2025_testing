```python
import numpy as np 
from qiskit import QuantumCircuit
from qiskit.circuit import Parameter
```

```python
qc = QuantumCircuit(5)
qc.h(4)
qc.cp( np.pi/2, 3, 4 )
qc.cp( np.pi/4, 2, 4 )
phi = Parameter(r"$\pi/2^{n\minus 1}$")
qc.cp( phi, 0, 4 )

qc.barrier()
qc.h(3)
qc.cp( np.pi/2, 2, 3 )
phi = Parameter(r"$\pi/2^{n\minus 2}$")
qc.cp( phi, 0, 3 )

qc.barrier()
qc.h(1)
qc.cp( np.pi/2, 0, 1 )

qc.barrier()
qc.h(0)

qc.draw('mpl', plot_barriers=False, filename='QFT.svg' )
```
