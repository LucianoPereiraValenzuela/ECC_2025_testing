```python
import numpy as np
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from ECC2025.testing import test_1
```

Durante el taller vimos la forma de implementar un estado GHZ para 3 qubits. Estos estados se pueden generalizar para un número arbitrario de qubits $n$ como
$$ |GHZ_n\rangle = \frac{1}{\sqrt{2^n}}\left( |0\dots0 \rangle + |1\cdots 1\rangle \right). $$
Este estado para $n=4$ qubits se prepara con el siguiente circuito:

```python
n_qubits = 4
qc_ghz = QuantumCircuit( n_qubits )
qc_ghz.h(0)
for j in range(n_qubits-1):
    qc_ghz.cx(j,j+1)

qc_ghz.draw('mpl')
```

```python
state = Statevector( qc_ghz )
state.draw("latex")
```

Notemos que este circuito esta formado por una Hadamard y una escalera de $n-1$ CNOTs, y que ninguna de estad puertas se esta realizando en paralelo. Esto quiere decir que la profundidad de este circuito es $n$.

```python
qc_ghz.depth()
```
