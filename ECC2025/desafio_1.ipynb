En 1989, los físicos Daniel Greenberger, Michael Horne y Anton Zeilinger hicieron un descubrimiento clave sobre el comportamiento de las partículas más pequeñas del universo. Desde los años 30, se sabía que dos partículas podían estar "entrelazadas", es decir, conectadas de una manera especial. Si algo le sucedía a una, la otra lo sentía de inmediato, incluso si estaban separadas por grandes distancias. Esta idea, conocida como la paradoja EPR, fue propuesta por Einstein y sus colegas.

Sin embargo, Greenberger, Horne y Zeilinger llevaron este concepto un paso más allá. Descubrieron que no solo dos, sino tres o más partículas podían estar entrelazadas, y que estas conexiones revelaban propiedades aún más sorprendentes. Este nuevo tipo de entrelazamiento, que llamaron estado GHZ, demostró que las partículas podían compartir una relación tan fuerte que no se podía explicar con las reglas normales de la física.

Este avance cambió nuestra comprensión del mundo cuántico, donde las partículas pueden estar misteriosamente conectadas, desafiando lo que creíamos posible.

https://arxiv.org/abs/0712.0921

Durante el taller, aprendimos a crear un estado GHZ utilizando tres partículas cuánticas, conocidas como qubits. Este tipo de estado puede extenderse a un número mayor de qubits, representado por $n$, de la siguiente manera:
$$ |GHZ_n\rangle = \frac{1}{\sqrt{2^n}}\left( |0\dots0 \rangle + |1\cdots 1\rangle \right). $$
Por ejemplo, si queremos preparar un estado GHZ para cuatro qubits ($n=4$), podemos hacerlo utilizando el siguiente circuito cuántico:

```python
n_qubits = 4
qc_ghz = QuantumCircuit( n_qubits )
qc_ghz.h(0)
for j in range(n_qubits-1):
    qc_ghz.cx(j,j+1)

qc_ghz.draw('mpl')
```

```python
state = Statevector( qc_ghz )
state.draw("latex")
```

En este circuito, se aplica una puerta Hadamard al primer qubit, seguida de una serie de $n−1$ puertas CNOT en forma de escalera. Ninguna de estas operaciones se realiza en paralelo, lo que significa que la profundidad del circuito es $n$.

Este diseño refleja cómo se establecen las conexiones entre los qubits paso a paso, profundizando el entrelazamiento cuántico de manera gradual a medida que se añaden más qubits al sistema.

```python
qc_ghz.depth()
```
