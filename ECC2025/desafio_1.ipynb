En la dcada de 1930, el fsico Albert Einstein, junto con sus colegas Podolsky y Rosen, formul la famosa paradoja EPR (Einstein-Podolsky-Rosen), un experimento mental que desafiaba la comprensin tradicional de la mecnica cuntica. En esta propuesta, argumentaron que dos partculas podan estar "entrelazadas", lo que implicaba que el estado de una partcula estaba intrnsecamente relacionado con el estado de la otra, sin importar la distancia que las separara. Esta nocin desafiaba el principio de localidad, sugiriendo que una accin en una partcula podra influir instantneamente en la otra, lo que Einstein describa como una "accin fantasmal a distancia". La paradoja EPR ha sido fundamental en el estudio de la fsica cuntica, impulsando numerosas investigaciones sobre las interacciones y conexiones entre partculas. Para una visin ms profunda de este tema, se puede consultar el artculo original de Einstein y sus colegas en [Physical Review](https://cds.cern.ch/record/405662/files/PhysRev.47.777.pdf).

Avanzando varias dcadas, en 1989, los fsicos Daniel Greenberger, Michael Horne y Anton Zeilinger ampliaron el concepto de entrelazamiento al demostrar que no solo dos partculas podan estar correlacionadas, sino que configuraciones ms complejas que involucraban tres o ms partculas tambin podan ser entrelazadas. Esta innovacin dio origen al estado GHZ (Greenberger-Horne-Zeilinger), una forma de entrelazamiento an ms rica y sorprendente. A diferencia del entrelazamiento de pares de partculas, donde la relacin se limita a dos, el estado GHZ revela cmo mltiples partculas pueden estar interconectadas de tal manera que sus estados son completamente dependientes unos de otros. Este avance no solo desafi las nociones tradicionales de la mecnica cuntica, sino que tambin proporcion un marco conceptual ms amplio para entender las interacciones cunticas. Para ms detalles, se puede consultar el trabajo de Greenberger, Horne y Zeilinger en el artculo disponible en [arXiv](https://arxiv.org/abs/0712.0921). La importancia de la investigacin en entrelazamiento cuntico fue reconocida con el Premio Nobel de Fsica 2022, que se otorg a Alain Aspect, John F. Clauser y Anton Zeilinger por sus experimentos que verificaron las predicciones de la mecnica cuntica en sistemas entrelazados; ms informacin se puede consultar en [Nobel Prize](https://www.nobelprize.org/prizes/physics/2022/summary/).

Matemticamente, el estado GHZ de $ n $ qubits se expresa como:

$$|GHZ_n\rangle = \frac{1}{\sqrt{2^n}}\left( |0\dots0\rangle + |1\cdots1\rangle \right).$$

Esto indica que el estado puede ser simultneamente $ |0\rangle $ en todos los qubits o $ |1\rangle $ en todos ellos. As, al medir uno de los qubits, se puede predecir con certeza el resultado de los dems, independientemente de la distancia que los separa. Por ejemplo, en un estado GHZ de cuatro qubits ($ n=4 $), la expresin se convierte en:

$$|GHZ_4\rangle = \frac{1}{\sqrt{16}}\left( |0000\rangle + |1111\rangle \right).$$

Esto significa que si se mide un qubit y se encuentra en el estado $ |0\rangle $, todos los otros qubits tambin estarn en $ |0\rangle $; de igual manera, si se mide un qubit y se encuentra en $ |1\rangle $, los dems qubits estarn en $ |1\rangle $. Este fenmeno ilustra la naturaleza no clsica del entrelazamiento cuntico, donde las partculas estn correlacionadas de tal manera que su comportamiento no puede describirse de forma independiente.


En la prctica para implementar un estado GHZ , utilizamos un circuito cuntico que establece estas interconexiones. Durante el taller, aprendimos a crear un estado GHZ utilizando qubits. El circuito diseado para preparar un estado GHZ de cuatro qubits ($ n=4 $) se construye de la siguiente manera:


```python
n_qubits = 4
qc_ghz = QuantumCircuit( n_qubits )
qc_ghz.h(0)
for j in range(n_qubits-1):
    qc_ghz.cx(j,j+1)

qc_ghz.draw('mpl')
```

En este circuito, se aplica una puerta Hadamard al primer qubit, colocndolo en una superposicin de $ |0\rangle $ y $ |1\rangle $. Luego, se utilizan puertas CNOT en forma de escalera para entrelazar los qubits restantes. Este proceso secuencial establece las conexiones entre los qubits, asegurando que todos compartan el estado GHZ.

Este diseo refleja cmo se establecen las conexiones entre los qubits paso a paso, profundizando el entrelazamiento cuntico a medida que se aaden ms qubits al sistema. La profundidad del circuito resultante es $ n $, dado que cada puerta CNOT se aplica de manera secuencial.

```python
qc_ghz.depth()
```

Este enfoque es beneficioso porque es lineal en el nmero de qubits. Sin embargo, existe una forma alternativa de implementar estados GHZ generalizados cuya profundidad es incluso menor que $ n $.

En el desafo anterior, exploramos la forma ms eficiente de implementar un estado GHZ en $4$ qubits. Sin embargo, esta estrategia puede no ser siempre la ms inteligente. Esto se debe a que los computadores cunticos tienen conectividad limitada, es decir, no todos los qubits estn conectados entre s. Tomemos como ejemplo el siguiente dispositivo cuntico de IBM de $5$ qubits:

burlington.jpg

En esta imagen, podemos observar que el qubit $0$ est conectado nicamente al qubit $1$, mientras que el qubit $1$ tiene conexiones con los qubits $0, 2$ y $3$. Si deseamos implementar un estado GHZ utilizando los qubits $0, 1, 2$ y $3$ de este dispositivo, notamos que los qubits $2$ y $3$ no estn directamente conectados. Esto nos obliga a descomponer la ltima puerta CNOT en una secuencia de $4$ puertas CNOT:

```python
qc_ghz_device = QuantumCircuit( n_qubits )
qc_ghz_device.h(0)
for j in range(n_qubits-2):
    qc_ghz_device.cx(j,j+1)

qc_ghz_device.cx(2,1)
qc_ghz_device.cx(1,2)
qc_ghz_device.cx(2,1)
qc_ghz_device.cx(1,3)

qc_ghz_device.draw('mpl')
```

Una vez construido el circuito, es crucial asegurarse de que realmente implementa el estado GHZ de manera correcta. Para lograrlo, podemos representar el vector de estado resultante:

```python
state = Statevector( qc_ghz_device )
state.draw("latex")
```

Esta visualizacin nos permite confirmar que el circuito ha generado el estado entrelazado esperado. Sin embargo, es importante considerar que la profundidad del circuito ha aumentado debido a la necesidad de utilizar ms puertas cunticas. Esto puede impactar negativamente la calidad de la ejecucin en un computador cuntico real:

```python
qc_ghz_device.depth()
```

Este incremento en la profundidad significa que la ejecucin de este circuito en un entorno real podra resultar en un rendimiento inferior, ya que ms puertas cunticas implican mayor susceptibilidad al ruido y a errores operativos. Por lo tanto, es esencial no solo validar el diseo del circuito, sino tambin simular su comportamiento en condiciones que imiten un dispositivo cuntico real.

Para ilustrar cmo la calidad de la implementacin se ve afectada por la profundidad del circuito y el ruido presente en el dispositivo, realizamos una simulacin que considera un modelo realista del dispositivo cuntico. Utilizamos un backend simulado que incorpora caractersticas de ruido tpicas de un entorno cuntico real:

```python
device_backend = FakeDevice()
simulator_noise = AerSimulator.from_backend(device_backend)
qc_ghz_device_measured = qc_ghz_device.copy()
qc_ghz_device_measured.measure_all()
qc_ghz_device_measured =  transpile( qc_ghz_device_measured, device_backend, optimization_level=0 )
counts_device = simulator_noise.run( qc_ghz_device_measured ).result().get_counts()

counts_ideal = { '0000':500, '1111':500  }
plot_histogram( [counts_ideal, counts_device], legend=['Ideal','Ruidoso'] )
```

En esta simulacin, comparamos los resultados ideales con los obtenidos en un entorno ruidoso, lo que nos permite visualizar cmo la calidad de la implementacin se ve afectada por la profundidad del circuito y el ruido presente en el dispositivo. Esta evaluacin es esencial para entender las limitaciones de los circuitos cunticos en situaciones prcticas, ya que el ruido puede distorsionar los resultados esperados y dificultar la implementacin de algoritmos cunticos complejos.
